```
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```
x = np.linspace(0, 10, 100)

fig = plt.figure(figsize=(10,6))
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--');
```

```
#create the first of two panels and set current axis
plt.figure(figsize=(10,6))

plt.subplot(2, 1, 1)  #roms, columns , panel number
plt.plot(x, np.sin(x))

#create the second panel and set current axis

plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x));
```

```
#Object-oriented interface
#lets do the same graph with oriented interface object

fig, ax = plt.subplots(2)


# Call plot() method on the appropriate object
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x));
```

## Simple Line Plots

```
#we start by creating a figure and an axes
plt.style.use('seaborn-whitegrid')

fig = plt.figure()
ax = plt.axes()
```

```
#now let's use the 1st example of figure
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0, 10, 100)
ax.plot(x, np.sin(x));
```

```
#we can use pylab, for more easy way
plt.figure(figsize=(10,5))
plt.plot(x, np.sin(x));
```

```
#Adjusting the Plot: Line Colors and Styles
#now let's do some configurations about colors and syles of lines

plt.figure(figsize=(14,7))
plt.plot(x, np.sin(x - 0), color = 'blue')
plt.plot(x, np.sin(x - 1), color = 'g') 
plt.plot(x, np.sin(x - 2), color = '#FFDD44')
plt.plot(x, np.sin(x - 3), color = '0.75')
plt.plot(x, np.sin(x - 4), color = (1.0, 0.2, 0.3))
plt.plot(x, np.sin(x - 5), color = 'chartreuse');


```

```
plt.figure(figsize=(17,8))
plt.plot(x, x + 0, linestyle = 'solid')
plt.plot(x, x + 1, linestyle = 'dashed')
plt.plot(x, x + 2, linestyle = 'dashdot')
plt.plot(x, x + 3, linestyle = 'dotted')
plt.plot(x, x + 4, linestyle = '-')
plt.plot(x, x + 5, linestyle = '--')
plt.plot(x, x + 6, linestyle = '-.')
plt.plot(x, x + 7, linestyle = ':');
```

```
#Adjusting the Plot: Axes Limits
#to adjust axis limits is just to use the plt.xlim() and plt.ylim() methods

plt.plot(x, np.sin(x))

plt.xlim(-2, 10)
plt.ylim(-2.5, 2.5);
```

```
#if you want to reverse, just reverse the numbers order
plt.plot(x, np.sin(x))
plt.xlim(10, -2)
plt.ylim(2.5, -2.5);
```

```
#to tighten the bounds the plot
plt.plot(x, np.sin(x))
plt.axis('tight');
```

```
#equal aspect ratio
plt.plot(x, np.sin(x))
plt.axis('equal');
```

```
#titles, axis labels, and simple legends (in axis)
plt.plot(x, np.sin(x))
plt.title('A cute curve ');
plt.xlabel('x')
plt.ylabel('sin(x)');

```

```
#we can to create a plot legend that labels each line type, for multiple lines
plt.figure(figsize=(12,6))
plt.plot(x, np.sin(x), '-g', label = 'sin(x)')
plt.plot(x, np.cos(x), ':b', label = 'cos(x)')
plt.axis('equal')

plt.legend();
```
