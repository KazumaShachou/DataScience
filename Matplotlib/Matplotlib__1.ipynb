## Simple Line Plots

```
#we start by creating a figure and an axes
plt.style.use('seaborn-whitegrid')

fig = plt.figure()
ax = plt.axes()
```

```
#now let's use the 1st example of figure
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0, 10, 100)
ax.plot(x, np.sin(x));
```

```
#we can use pylab, for more easy way
plt.figure(figsize=(10,5))
plt.plot(x, np.sin(x));
```

```
#Adjusting the Plot: Line Colors and Styles
#now let's do some configurations about colors and syles of lines

plt.figure(figsize=(14,7))
plt.plot(x, np.sin(x - 0), color = 'blue')
plt.plot(x, np.sin(x - 1), color = 'g') 
plt.plot(x, np.sin(x - 2), color = '#FFDD44')
plt.plot(x, np.sin(x - 3), color = '0.75')
plt.plot(x, np.sin(x - 4), color = (1.0, 0.2, 0.3))
plt.plot(x, np.sin(x - 5), color = 'chartreuse');


```

```
plt.figure(figsize=(17,8))
plt.plot(x, x + 0, linestyle = 'solid')
plt.plot(x, x + 1, linestyle = 'dashed')
plt.plot(x, x + 2, linestyle = 'dashdot')
plt.plot(x, x + 3, linestyle = 'dotted')
plt.plot(x, x + 4, linestyle = '-')
plt.plot(x, x + 5, linestyle = '--')
plt.plot(x, x + 6, linestyle = '-.')
plt.plot(x, x + 7, linestyle = ':');
```

```
#Adjusting the Plot: Axes Limits
#to adjust axis limits is just to use the plt.xlim() and plt.ylim() methods

plt.plot(x, np.sin(x))

plt.xlim(-2, 10)
plt.ylim(-2.5, 2.5);
```

```
#if you want to reverse, just reverse the numbers order
plt.plot(x, np.sin(x))
plt.xlim(10, -2)
plt.ylim(2.5, -2.5);
```

```
#to tighten the bounds the plot
plt.plot(x, np.sin(x))
plt.axis('tight');
```

```
#equal aspect ratio
plt.plot(x, np.sin(x))
plt.axis('equal');
```

```
#titles, axis labels, and simple legends (in axis)
plt.plot(x, np.sin(x))
plt.title('A cute curve ');
plt.xlabel('x')
plt.ylabel('sin(x)');

```

```
#we can to create a plot legend that labels each line type, for multiple lines
plt.figure(figsize=(12,6))
plt.plot(x, np.sin(x), '-g', label = 'sin(x)')
plt.plot(x, np.cos(x), ':b', label = 'cos(x)')
plt.axis('equal')

plt.legend();
```

## Simple Scatter Plots

```
#Scatter Plots with plt.plot
plt.figure(figsize=(10,5))
x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x, y, 'o', color = 'black');
```

```
#Markers
plt.figure(figsize=(13,6))
rng = np.random.RandomState(72)
for marker in ['o', '.' ,'-' ,'v', '+', '^', '<', '>', 's', 'd']:
  plt.plot(rng.rand(5), rng.rand(5), marker,
           label = "marker = {0}".format(marker))
  plt.legend(numpoints = 1)
  plt.xlim(0, 1.2);
```

```
#character codes can be used together with line and color codes to plot points along with a line connecting them
plt.plot(x, y, '-ok');   #k for black
```

```
plt.plot(x, y, '-<m')  #m for magenta
```

```
#for Additional keyword
plt.figure(figsize=(12,6))
plt.plot(x, y, '-o', color= 'black', markersize = 15, linewidth = 3, markerfacecolor = 'red', markeredgecolor = 'blue',markeredgewidth= 2)
plt.xlim(-0.1, 10);
plt.title('The snake-chan')
plt.xlabel('size')
plt.ylabel('IDK');
```

```
#we can make scartter plot using too plt.scatter
plt.scatter(x, y, marker='<');
```

```
#we can make more detail graph using scatter too
plt.figure(figsize=(14,6))
ks = np.random.RandomState(84)
x = ks.randn(150)
y = ks.randn(150)
colors = ks.rand(150)
sizes = 1000* rng.rand(150)

plt.scatter(x, y, c = colors, s = sizes, alpha = 0.4, cmap = 'inferno')
plt.colorbar();
```

```
#use the Iris data from Scikit-Learn, for see for example, type of petal, size measured
from sklearn.datasets import load_iris
plt.figure(figsize=(14,7))
iris = load_iris()
features = iris.data.T

plt.scatter(features[0], features[1], alpha = 0.7, s = 100*features[3],  c= iris.target, cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1]);
```

## Density and Contour Plots

```
#Visualizing a Three-Dimensional Function
def f(x,y):
  return np.sin(x) **15 + np.cos(10 + y * x) * np.cos(x)
```

```
x = np.linspace(0, 5, 100)
y = np.linspace(0, 5, 50)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```
#negative values are represented by dashed lines, and positive values by solid lines
plt.figure(figsize=(10,5))
plt.contour(X, Y, Z, colors = 'grey');
```

```
plt.figure(figsize=(10,5))
plt.contour(X, Y, Z, 20, cmap = 'plasma');
```

```
plt.figure(figsize=(10,5))
plt.contour(X, Y, Z, 20, cmap = 'plasma');
plt.colorbar();
```

```
plt.figure(figsize=(10,5))
plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower', cmap='plasma')
plt.colorbar()
plt.axis(aspect = 'image');
```

```
plt.figure(figsize=(12,7))
contours = plt.contour(X, Y, Z, 5, colors= 'black')
plt.clabel(contours, inline = True, fontsize = 11)

plt.imshow(Z, extent=[0, 5, 0, 5], origin= 'lower', cmap='plasma', alpha=0.6)
plt.colorbar();
```
